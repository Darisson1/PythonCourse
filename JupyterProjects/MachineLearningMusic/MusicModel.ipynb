```python
import pandas as pd
from sklearn.model_selection import train_test_split


music_data = pd.read_csv('music.csv')
# drop with par colums: new dataset without the column we choose. X (upper case) is the input?
X = music_data.drop(columns=['genre'])
X
# y (lower case) is the output
y = music_data['genre']
y

# split the data into 4 lists: Input for training and testing and output for training and testing
# the function train_test_split RANDOMLZ picks training and testdata of the data we give 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# new instance of DecisionTreeClassifier
model = DecisionTreeClassifier()
# train the model with the input X and output y. 
# In the tut there was only (X, y) but there was a warning and at stackoverflow was a fix with .values
# https://stackoverflow.com/questions/69326639/sklearn-warning-valid-feature-names-in-version-1-0
# edit: I think I have to run all cells. No error even without the .values
model.fit(X_train, y_train)

# prediction of the model: input a two-dimensional array. In the inner arrays: 21 year old male, 22 year old female
# predictions = model.predict([ [21, 1], [22, 0] ])

# edit: after separating the data into test and training data, we will give the model the test data:
predictions = model.predict(X_test)

#measure the accuracy with the function of the library: compare the output test data with the predictions of the model. between 0 and 1
score = accuracy_score(y_test, predictions)
score
```
